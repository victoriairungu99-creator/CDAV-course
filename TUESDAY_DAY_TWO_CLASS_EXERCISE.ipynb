# Create a simple program that asks a user for two values and returns the mathematical operations in a user friendly manner

```python
value_1=int(input("Enter number of choice"))
value_2=int(input("Enter number of choice"))
print("the sum of both values is", value_1+value_2)
print("the subtraction of both values is", value_1-value_2)
print("the division of both values is", value_1/value_2)
print("the floor division of both values is", value_1//value_2)
print("the modulus after division of both values is", value_1%value_2)
```
