# Exercise 72 on python workbook

```python
word=input("Enter word: ")
i=0
palindrome=True
while i<len(word)//2:
  if word[i]==word[len(word)-i-1]:
    i=i+1
  else:
    palindrome=False


if palindrome:
  print("The word is a palindrome")
else:
  print("The word is not a palindrome")





```

# Exercise 75 on python workbook

```python
m=int(input("Enter integer: "))
n=int(input("Enter integer: "))

if m<n:
  d=m
else:
  d=n

while m%d!=0 or n%d!=0:
    d=d-1
print("The GCD is",d)





```

# Exercise 70 on python workbook

```python
message=input("Enter message: ")
shift=int(input("Enter the shift amount: "))

result=""

for ch in message:
  if ch.isupper():
    new_char=chr((ord(ch)-ord(A)+shift))%26+ord('A')
    result+=new_char

  elif ch.islower():
    new_char=chr((ord(ch)-ord('a')+shift)%26+ord('a'))
    result+=new_char
  else:
    result+=ch

print("Shifted message:",result)






```
