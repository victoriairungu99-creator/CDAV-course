```python
negative = []
zeros = []
positive = []

entry = input("Enter an integer or press Enter to stop: ")

while entry != "":
    number = int(entry)

    if number < 0:
        negative.append(number)
    elif number == 0:
        zeros.append(number)
    else:
        positive.append(number)

    entry = input("Enter an integer or press Enter to stop: ")

for num in negative:
    print(num)

for num in zeros:
    print(num)

for num in positive:
    print(num)

```

Sets

-They are unordered- the items may not come out in the same order everytime in different outputs

They are mutable- the final set can be altered using .add , .remove and .discard

-The colection of items is unique(each item only features once in output)

-Their only functions are.add ,.remove and .discard(for.remove,the item has to exist in the list before or it wil  bring an eror but for .discard it is error free(fullproof) and it will just remove an output regardles of if it exists or not)

```python
#for list:
fruits=["apple","banana","mango","banana","plum"]
print(fruits)
```

```python
fruits_set={"apple","banana","mango","banana","plum"}
print(fruits_set)
```

```python
fruits_set.add("guava")
print(fruits_set)
```

```python
fruits_set.remove("apple")
print(fruits_set)
```

```python
fruits_set.remove("banana")
print(fruits_set)
```

```python
fruits_set.discard("apple")
print(fruits_set)
```
